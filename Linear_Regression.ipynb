```python
import pandas as pd 
import numpy as np 
import matpllotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df =pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.info()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Use square brackets and quotes for column names with spaces/special characters
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')

# Add labels (optional)
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income Over Time')
plt.show()
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['per capita income (US$)'] = pd.to_numeric(df['per capita income (US$)'], errors='coerce')
```

```python
X = df['year'].values.reshape(-1, 1)  # Reshape to 2D array
y = df['per capita income (US$)'].values

reg = linear_model.LinearRegression()
reg.fit(X, y)
```
