# Transforming Object to Integer  
- transforming Yes to 1 and No to 0


```python
# Check if the column exists and drop it
if 'furnishingstatus' in df.columns:
    df.drop(columns=['furnishingstatus'], inplace=True)
```

```python
df
```

```python
# List of columns to convert
bool_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Convert "Yes" to 1 and "No" to 0 in the specified columns
for column in bool_columns:
    if column in df.columns:  # Ensure the column exists
        # Handle missing values (e.g., replace NaN with 0 or any default value)
        df[column] = df[column].map({'Yes': 1, 'No': 0})
        df[column] = df[column].fillna(0).astype(int)  # Fill NaN with 0 and cast to integer
```

```python
df =pd.get_dummies(df,drop_first=True)
df.head()
```

```python
X =df.drop('price',axis =1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X,train_y)
```

```python
reg.score(test_X,test_y)
```

```python
reg.score(train_X,train_y)
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
lasso_reg.score(train_X,train_y)
```

# Which is a Better Model 
We see that the score of Lasso is Better than Linear Regression
